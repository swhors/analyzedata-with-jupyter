```python
# db configuration
db_host="localhost"
db_port=3306
db_user="tester"
db_passwd="aq1234"
db_db="finance"
charset="utf8"
```

```python
# import library
import pymysql
import os
import FinanceDataReader as fdr
conn = None
```

```python
# db connect
def db_init():
    global conn
    conn = pymysql.connect(host=db_host, user=db_user, password=db_passwd, db=db_db, charset=charset)
    return conn
```

```python
# close connection
def db_fint(conn):
    if conn != None:
        conn.close()
```

```python
# select data
def get_all(conn) -> []:
    cur = conn.cursor()
    stocks = []
    cur.execute("select * from finance.stock_list;")
    rows = cur.fetchall()
    print(f'fetched data = {len(rows)}')
    for row in rows:
        print(row)
    # if len(row) > 0:
    cur.close()
    return stocks
```

```python
def put_datas(conn, stocks):
    cur = conn.cursor()
    for stock in stocks:
        cmd = f"insert into finance.stock_list values({str(stock)})"
        print(f"cmd = {cmd}")
        cur.execute(cmd)
    conn.commit()
    cur.close()
```

```python
def readers_2_stocks():
    from stock import Stock
    all_stocks = []
    korea_markets = ['KOSPI-DESC', 'KOSDAQ-DESC', 'KONEX-DESC'] # KOSPI: 940 종목, KOSDAQ: 1,597 종목, KONEX: 126 종목
    for market in korea_markets:
        stocks = fdr.StockListing(market)
        stock_dict = stocks.to_dict('records')
        for d in stock_dict:
            #print(f'{type(d)}, {d}')
            all_stocks.append(Stock.dict_2_stock(d))
        # all_stocks.extend(stocks)
        #print(f'{market}')
        #stock = Stock(market,"","","","","","","","")
        #print(f'{stock}')
    return all_stocks
```
